```python
import pandas as pd
import matplotlib.pyplot as plt
import sys
import numpy as np
```

```python
df_0=pd.read_csv('pos0_merged.csv',names=["matches_0","strain","threshold_0"])
df_1=pd.read_csv('pos1_merged.csv',names=["matches_1","strain","threshold_1"])
df_2=pd.read_csv('pos2_merged.csv',names=["matches_2","strain","threshold_2"])
df_3=pd.read_csv('pos3_merged.csv',names=["matches_3","strain","threshold_3"])
df_4=pd.read_csv('pos4_merged.csv',names=["matches_4","strain","threshold_4"])
df_11=pd.read_csv('pos-1_merged.csv',names=["matches_11","strain","threshold_11"])
df_22=pd.read_csv('pos-2_merged.csv',names=["matches_22","strain","threshold_22"])
df_33=pd.read_csv('pos-3_merged.csv',names=["matches_33","strain","threshold_33"])
df_44=pd.read_csv('pos-4_merged.csv',names=["matches_44","strain","threshold_44"])
total=pd.read_csv('total_chr.csv',names=["total"])
index=pd.read_csv('sample.txt',names=["tissue"])
```

```python
df_merged=pd.concat([df_0,df_1,df_2,df_3,df_4,df_11,df_22,df_33,df_44,total,index], axis= 1)
```

```python
df_merged=df_merged.drop(columns=['strain'])
```

```python
df_merged["fract_0"]=df_merged["matches_0"]/df_merged["total"]
df_merged["fract_1"]=df_merged["matches_1"]/df_merged["total"]
df_merged["fract_2"]=df_merged["matches_2"]/df_merged["total"]
df_merged["fract_3"]=df_merged["matches_3"]/df_merged["total"]
df_merged["fract_4"]=df_merged["matches_4"]/df_merged["total"]
df_merged["fract_11"]=df_merged["matches_11"]/df_merged["total"]
df_merged["fract_22"]=df_merged["matches_22"]/df_merged["total"]
df_merged["fract_33"]=df_merged["matches_33"]/df_merged["total"]
df_merged["fract_44"]=df_merged["matches_44"]/df_merged["total"]
```

```python
import seaborn as sns
sns.scatterplot(x=df_merged["threshold_0"], y=df_merged["fract_0"], hue=df_merged["tissue"])
sns.scatterplot(x=df_merged["threshold_1"], y=df_merged["fract_1"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_2"], y=df_merged["fract_2"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_3"], y=df_merged["fract_3"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_4"], y=df_merged["fract_4"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_11"], y=df_merged["fract_11"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_22"], y=df_merged["fract_22"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_33"], y=df_merged["fract_33"], hue=df_merged["tissue"],legend=False)
sns.scatterplot(x=df_merged["threshold_44"], y=df_merged["fract_44"], hue=df_merged["tissue"],legend=False)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)
plt.xlabel("Threshold (bp)")
plt.ylabel("Fraction overlap between differential methylation and variants")
plt.savefig(sys.argv[1],bbox_inches='tight',dpi=300)
```
